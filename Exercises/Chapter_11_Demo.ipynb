## Pieplot

```r
install.packages('plotrix')
```

```r
library(plotrix)
```

```r
x = c(46, 37, 35, 53)
labels = c('10A1', '10A2', '10A3', '10A4')
# png(file = 'Du_lieu_new/pie.png')
pie(x, labels, col=rainbow(length(x)), clockwise=TRUE)
# dev.off()
```

```r
piepercent = round(100*x/sum(x), 1)
pie(x, labels = paste(labels,'-', piepercent, '%'), main='Classes pie chart', col=rainbow(length(x)), clockwise=TRUE)
legend('topright', labels, cex = 0.8, fill = rainbow(length(x)))
```

```r
pie3D(x, labels = labels, main='Classes pie chart 3D', col=rainbow(length(x)))
```

```r
bears = c(8, 54, 93, 116, 137)
dolphins = c(150, 77, 32, 11, 6)
whales = c(80, 54, 100, 76, 93)

g_range = range(0, bears, dolphins, whales)
plot(bears, type='o', col='red', xlab='Year', ylab='Numbers', main='Numbers of animal',
    xaxt='n', ylim=g_range)

axis(1, at=1:5, labels=2017:2021)

lines(dolphins, type='o', col='blue')
lines(whales, type='o', col='blue')
legend('top', inset=0.01, c('bears', 'dolphins', 'whales'), cex=0.8, fill=c('red', 'blue', 'green'))
```

## Wordcloud

```r
library("tm")
library("SnowballC")
library("wordcloud")
library("RColorBrewer")
```

```r
text = readLines('Du_lieu/introduction_to_DS_MK.txt')
docs = Corpus(VectorSource(text))
```

```r
inspect(docs)
```

```r
toSpace = content_transformer(function(x, pattern) gsub(pattern, ' ', x))
docs = tm_map(docs, toSpace, '/')
docs = tm_map(docs, toSpace, '@')
docs = tm_map(docs, toSpace, '\\|')
```

```r
stopwords('english')
```

```r
docs = tm_map(docs, content_transformer(tolower))
docs = tm_map(docs, removeWords, stopwords('english'))
docs = tm_map(docs, removeNumbers)
docs = tm_map(docs, removeWords, c('then', 'just', 'caldwell'))
docs = tm_map(docs, removePunctuation)
docs = tm_map(docs, stripWhitespace)
```

```r
dtm = TermDocumentMatrix(docs)
m = as.matrix(dtm)
v = sort(rowSums(m), decreasing=TRUE)
d = data.frame(word=names(v), freq=v)
head(d, 10)
```

```r
set.seed(42)
wordcloud(words = d$word, freq = d$freq, 
          min.freq=1, max.words=200, 
          random.order=FALSE, rot.per=0.35,
         colors=brewer.pal(8, 'Dark2'))
```

```r
findFreqTerms(dtm, lowfreq=3)
```

```r
findAssocs(dtm, terms='data', corlimit=0.5)
```

```r
barplot(d[1:10, ]$freq, las=2, names.arg=d[1:10, ]$word,
       col='lightblue', main='Most frequent words',
       ylab = 'Word frequencies')
```

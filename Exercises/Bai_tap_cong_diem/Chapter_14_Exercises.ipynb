```r
model = HoltWinters(df_ts, seasonal = 'multiplicative')
model
```

Nhn xt:
- Alpha = 0.57 trung bnh: gi tr hin ti c ph thuc 1 t vo d liu gn y
- Beta = 0.01 rt thp:  dc ca trend l c lp trong tng thi im
- Gamma = 0.54 trung bnh: nh hng ca yu t ma ti thi im hin ta c ph thuc vo gi tr gn nht

```r
sse = model$SSE
sse
```

```r
mse = sse/length(df_ts)
mse
```

```r
rmse = sqrt(mse)
rmse
```

```r
plot(model, main='Births observer vs forecast',
     xlab = 'Year', ylab = 'New born')
```

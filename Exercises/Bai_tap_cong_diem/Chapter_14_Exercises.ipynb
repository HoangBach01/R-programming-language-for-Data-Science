# Ex 2

```r
df = scan('../Du_lieu/mybirths.dat.txt', skip=3)
df
```

```r
df_ts = ts(df, start=c(1946, 1), frequency=12)
df_ts
```

```r
plot(df_ts, main='New born per month in New York \n from 01/1946 to 12/1959',
    xlab='Year', ylab='New Born')
```

Nhận xét:
- Từ năm 1949, số trẻ sinh ra hàng tháng có xu hướng tăng qua các năm.

```r
comp = decompose(df_ts)
comp$seasonal
```

Nhận xét: có tính seasonal, tăng vào mùa hè, giảm về mua thu đông, tăng lại vào đầu năm

```r
comp$trend
```

Nhận xét: có trend tăng

```r
comp$random
```

```r
plot(comp)
```

```r
model = HoltWinters(df_ts, seasonal = 'multiplicative')
model
```

Nhận xét:
- Alpha = 0.57 trung bình: giá trị hiện tại có phụ thuộc 1 ít vào dữ liệu gần đây
- Beta = 0.01 rất thấp: độ dốc của trend là độc lập trong từng thời điểm
- Gamma = 0.54 trung bình: ảnh hưởng của yếu tố mùa tại thời điểm hiện taị có phụ thuộc vào giá trị gần nhất

```r
sse = model$SSE
sse
```

```r
mse = sse/length(df_ts)
mse
```

```r
rmse = sqrt(mse)
rmse
```

```r
plot(model, main='Births observer vs forecast',
     xlab = 'Year', ylab = 'New born')
```

```r
forecast.new = forecast:::forecast.HoltWinters(model, h = 6)
forecast.new
```

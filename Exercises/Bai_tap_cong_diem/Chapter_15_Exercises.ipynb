# Ex 2

```r
sample = rnorm(50, mean=156, sd=4.6)
sample = ceiling(sample*2) /2
sample = sample[sample <= 180]
sample = sample[sample >= 140]
print(sample)
```

```r
range = seq(140, 180, by=0.5)
prob = dnorm(range, mean=156, sd=4.6)
print(prob)
```

```r
hist(sample)
```

```r
plot(range, prob)
```

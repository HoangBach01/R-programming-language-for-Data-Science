# Normal - Binormal Distribution

## 1. IQ

### a. Tỷ lệ phần trăm người có IQ nhỏ hơn 125

```r
percent = pnorm(125, mean = 100, sd = 15)*100
print(paste('Tỷ lệ phần trăm người có IQ nhỏ hơn 125 là:',round(percent,2), '%'))
```

### b. Tỷ lệ phần trăm người có IQ lớn hơn 110

```r
percent = 100-pnorm(110, mean = 100, sd = 15)*100
print(paste('Tỷ lệ phần trăm người có IQ lớn hơn 110 là:',round(percent,2), '%'))
```

### c. Tỷ lệ phần trăm người có IQ trong khoảng từ 110 và 125

```r
percent = pnorm(125, mean = 100, sd = 15)*100 - pnorm(110, mean = 100, sd = 15)*100
print(paste('Tỷ lệ phần trăm người có IQ trong khoảng 110 và 125 là:',round(percent,2), '%'))
```

## 2. Xúc xắc 6 mặt

### a. Xác xuất để có được 2 lần mặt 4 nút trong 5 lần đổ xúc xắc

```r
percent = dbinom(2, 5, 1/6)*100
print(paste('Xác xuất để có được 2 lần mặt 4 nút trong 5 lần đổ xúc xắc là:',round(percent,2), '%'))
```

### b. Số mặt 4 nút khi có xác xuất 25% xuất hiện khi một xúc xắc được đổ 50 lần

Nếu mặt 4 nút có xác suất 25% xuất hiện, thì khi đổ 50 lần, trung bình sẽ có 25% x 50 = 12.5 (lần) xuất hiện mặt xúc xắc 4 nút đó.

Biểu diễn phân phối việc xuất hiện mặt 4 bút trên bằng đồ thị như sau

```r
x = seq(0, 50, by = 1)
y = dbinom(x, size = 50, prob = 0.25)
plot(x, y)
```

- Trục x là số lần xuất hiện, trục y là xác xuất xuất hiện của số lần x đó.
- Có thể dễ dàng thấy được phần lớn xác xuất sẽ cho ra số lần xuất hiện mặt 4 nút từ 5-20 lần, với số lần có xác xuất cao nhất là 12.5 lần

```r
percent = pbinom(20, 50, 0.25)*100 - pbinom(5, 50, 0.25)*100
print(paste('Xác xuất sẽ cho ra số lần xuất hiện mặt 4 nút từ 5-20 lần là:',round(percent,2), '%'))
```
